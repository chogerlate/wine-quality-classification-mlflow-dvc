```python
import pandas as pd

def load_wine_quality_data(data_path: str) -> pd.DataFrame:
    """
    Summary: Load the wine quality dataset from the specified directory.

    Args:
        data_path (str): The path to the directory containing the wine quality dataset.

    Returns:
        pd.DataFrame: A DataFrame containing the wine quality data.
    """
    file_path = f"{data_path}/WineQT.csv"
    data = pd.read_csv(file_path)
    return data

# Load the dataset
wine_quality_data = load_wine_quality_data("../data")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_quality_distribution(data: pd.DataFrame) -> None:
    """
    Summary: Plot the distribution of wine quality classes.

    Args:
        data (pd.DataFrame): A DataFrame containing the wine quality data.

    Returns:
        None
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(x='quality', data=data, palette='viridis')
    plt.title('Distribution of Wine Quality Classes')
    plt.xlabel('Quality')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.grid(axis='y')
    plt.show()

# Plot the distribution of wine quality classes
plot_quality_distribution(wine_quality_data)
```
